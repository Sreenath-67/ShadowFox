# Language Model Exploration and Analysis

## Objective
This project explores the capabilities of a pre-trained Language Model (GPT-2) using Hugging Face Transformers.

## Goals
- Analyze contextual understanding
- Evaluate creativity
- Test domain adaptability
- Visualize performance
- Identify strengths and limitations

```python
from transformers import pipeline
import matplotlib.pyplot as plt
from nltk.translate.bleu_score import sentence_bleu
import pandas as pd
```

```python
generator = pipeline("text-generation", model="gpt2")
def generate_text(prompt, max_length=50):
    output = generator(prompt, max_length=max_length, num_return_sequences=1)
    return output[0]['generated_text']
```

## Test Scenarios

We evaluate the model using:
1. Context Understanding
2. Creativity
3. Domain Adaptability

```python
context_prompts = [
    "Ravi went to the market and bought apples. What did Ravi buy?",
    "The sky is blue because of scattering of light. Why is the sky blue?"
]

creative_prompts = [
    "Write a sci-fi story about AI taking over:",
    "Describe a world where robots rule humans:"
]

domain_prompts = {
    "Medical": "Symptoms of diabetes include",
    "Coding": "Write a Python function to calculate factorial",
    "General": "Explain climate change"
}
```

## Context Understanding Analysis

This section evaluates whether the LM retains and uses contextual information.

```python
context_results = []

for prompt in context_prompts:
    output = generate_text(prompt)
    print("PROMPT:", prompt)
    print("OUTPUT:", output)
    print("-" * 60)
    context_results.append(output)
```

## Creativity Analysis

We test how well the model generates imaginative and coherent text.

```python
creative_results = []

for prompt in creative_prompts:
    output = generate_text(prompt, max_length=100)
    print("PROMPT:", prompt)
    print("OUTPUT:", output)
    print("-" * 60)
    creative_results.append(output)
```

## Domain Adaptability

We test how the model performs across different domains.

```python
domain_results = {}

for domain, prompt in domain_prompts.items():
    output = generate_text(prompt)
    print(domain.upper(), ":", output)
    print("-" * 60)
    domain_results[domain] = output
```

## Evaluation Metrics

We evaluate:
- Response Length
- BLEU Score (basic quality comparison)

```python
def compute_bleu(reference, candidate):
    reference = [reference.split()]
    candidate = candidate.split()
    return sentence_bleu(reference, candidate)

def response_length(text):
    return len(text.split())
```

```python
all_outputs = context_results + creative_results + list(domain_results.values())

lengths = [response_length(text) for text in all_outputs]

print("Response Lengths:", lengths)
```

## Visualization of Results

```python
plt.hist(lengths)
plt.title("Response Length Distribution")
plt.xlabel("Length")
plt.ylabel("Frequency")
plt.show()
```
